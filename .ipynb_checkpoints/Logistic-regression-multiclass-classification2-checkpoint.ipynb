```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
iris['species'].unique()
```

```python
iris['species'].replace({'Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'}, inplace = True)
```

```python
iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(iris[['sepal_length','SepalWidthCm','PetalLengthCm','PetalWidthCm']],iris['Species'],test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
x_test   # Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'
```

```python
lr.score(x_test,y_test)
```
